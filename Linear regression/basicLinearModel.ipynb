```python
# # Import the necessary libraries
# import seaborn as sns
# import pandas as pd
# import numpy as np
# import matplotlib.pyplot as plt
# from sklearn.linear_model import LinearRegression
# from sklearn.metrics import mean_squared_error, r2_score

# # Load Anscombe's quartet dataset
# anscombe = sns.load_dataset("anscombe")

# # Separate the dataset into four subsets (I, II, III, IV)
# anscombe_I = anscombe[anscombe["dataset"] == "I"]
# anscombe_II = anscombe[anscombe["dataset"] == "II"]
# anscombe_III = anscombe[anscombe["dataset"] == "III"]
# anscombe_IV = anscombe[anscombe["dataset"] == "IV"]

# # Loop through the four datasets and perform linear regression and analysis
# datasets = [anscombe_I, anscombe_II, anscombe_III, anscombe_IV]

# for i, dataset in enumerate(datasets):
#     # Define X and Y
#     X = dataset["x"]
#     y = dataset["y"]

#     # Perform linear regression
#     model = LinearRegression()
#     model.fit(X.values.reshape(-1, 1), y)

#     # Calculate statistics
#     min_x = np.min(X)
#     max_x = np.max(X)
#     mean_x = np.mean(X)
#     min_y = np.min(y)
#     max_y = np.max(y)
#     mean_y = np.mean(y)

#     # Print statistics
#     print(f"Dataset {i + 1} Statistics:")
#     print(f"X - Min: {min_x}, Max: {max_x}, Mean: {mean_x}")
#     print(f"Y - Min: {min_y}, Max: {max_y}, Mean: {mean_y}")

#     # Print regression parameters
#     print(f"Linear Regression - Slope: {model.coef_[0]}, Intercept: {model.intercept_}")

#     # Make predictions
#     y_pred = model.predict(X.values.reshape(-1, 1))

#     # Calculate accuracy measures
#     r2 = r2_score(y, y_pred)
#     mse = mean_squared_error(y, y_pred)

#     # Print accuracy measures
#     print(f"R^2: {r2}, MSE: {mse}")

#     # Create diagnostic plots
#     plt.figure(figsize=(8, 6))
#     plt.scatter(X, y, label="Data")
#     plt.plot(X, y_pred, color='red', label="Linear Regression")
#     plt.title(f"Dataset {i + 1} - Linear Regression")
#     plt.xlabel("X")
#     plt.ylabel("Y")
#     plt.legend()
#     plt.show()

#     #


# Import the necessary libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load Anscombe's quartet dataset
anscombe = sns.load_dataset("anscombe")

# Separate the dataset into four subsets (I, II, III, IV)
anscombe_I = anscombe[anscombe["dataset"] == "I"]
anscombe_II = anscombe[anscombe["dataset"] == "II"]
anscombe_III = anscombe[anscombe["dataset"] == "III"]
anscombe_IV = anscombe[anscombe["dataset"] == "IV"]

# Create a list of datasets for iteration
datasets = [anscombe_I, anscombe_II, anscombe_III, anscombe_IV]

for i, dataset in enumerate(datasets):
    # Calculate statistics using the describe method
    dataset_statistics = dataset.describe()

    # Print the dataset statistics
    print(f"Dataset {i + 1} Statistics:")
    print(dataset_statistics)

    # Define X and Y
    X = dataset["x"]
    y = dataset["y"]

    # Perform linear regression
    model = LinearRegression()
    model.fit(X.values.reshape(-1, 1), y)

    # Print regression parameters
    print(f"Linear Regression - Slope: {model.coef_[0]}, Intercept: {model.intercept_}")

    # Make predictions
    y_pred = model.predict(X.values.reshape(-1, 1))

    # Calculate accuracy measures
    r2 = r2_score(y, y_pred)
    mse = mean_squared_error(y, y_pred)

    # Print accuracy measures
    print(f"R^2: {r2}, MSE: {mse}")

    # Create diagnostic plots
    plt.figure(figsize=(8, 6))
    plt.scatter(X, y, label="Data")
    plt.plot(X, y_pred, color='red', label="Linear Regression")
    plt.title(f"Dataset {i + 1} - Linear Regression")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.show()

# from the plots we can see that linear regression works fine for dataset 1 and 3.
# but for dataset 2 the data is represented with polinomial of degree 2, and linear model does not fit well for that data
# dataset 4 contains fixed constant for X, and an outlier, and here linear model does not fit well

```
